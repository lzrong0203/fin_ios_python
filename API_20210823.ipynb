```python
import requests
import json

url = "https://api.github.com"
r = requests.get(url)
api_dict = json.loads(r.text)
list_keys = list(api_dict.keys())
```

```python
repos_search = []
for i in list_keys:
    if i.startswith("repos"):
        repos_search.append(i)
```

```python
api_dict[repos_search[1]]
```

```python
list1 = api_dict[repos_search[1]].split("&")
url = list1[0].rstrip("{")
queryURL = url.format(query="python") + "&sort=stars"

git_python = requests.get(queryURL)
```

```python
reponse_dict = git_python.json()
```

```python
for k in reponse_dict.keys():
    print(k, ": ", end="")
    print(reponse_dict[k])
```

```python
repo_dict = reponse_dict["items"]
```

```python
repo_dict
```

```python
queryURL
```

```python
from plotly import offline
```

```python
repo_links, stars, label = [], [], []

for repo in repo_dict:
    repo_name = repo["name"]
    repo_url = repo["html_url"]
    stars.append(repo["stargazers_count"])
    repo_link = f"<a href='{repo_url}'>{repo_name}</a>"
    repo_links.append(repo_link)
```

```python
data = [{"type": "bar",
         "x": repo_links,
         "y": stars,
         "marker": {"color": 'rgb(60, 60, 150)'}}]
layouts = {"title": "Most stars Python Project"}
fig = {"data": data, "layout": layouts}
offline.plot(fig, filename="python_git.html")
```
